```python
import matplotlib.pyplot as plt

## Plot the data
fig, ax = plt.subplots(figsize=(10,6))

## Customize the plot
scatter = ax.scatter(x=df["sepal_length"],
                   y=df["sepal_width"],
                   c=df["petal_length"])

## Add a legend
ax.set(title="IRIS",
       xlabel="sepal_length",
       ylabel="sepal_width")

## Add a horixontal line or meanline
ax.legend(*scatter.legend_elements(), title="petal_length")

## To change the format of line
ax.axhline(df["sepal_width"].mean(),
           linestyle="--")
```

```python

Q1 = df["sepal_length"].quantile(0.25)
Q2 = df["sepal_length"].quantile(0.50)
Q3 = df["sepal_length"].quantile(0.75)

IQR = Q3 - Q1

print("Q1: ", Q1)
print("Q2: ", Q2)
print("Q3: ", Q3)

print("IQR: ",IQR)
```
