#### Types of missing vals

1. Missing vals completely at random (MCAR)

        Ex: A survey respondent skips a question because they misread it.
        
###

2. NULL vals are actually dependent on other vals/ Random vals aren't completely missing

        Ex: A survey respondent skips a question because they don't want to answer it.

###

3. Missing vals are dependent on the missing vals themselves

        Ex: A survey respondent skips a question because they don't want to answer it, and the question is about their income.
