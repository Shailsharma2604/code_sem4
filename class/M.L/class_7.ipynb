# Generalized FÎ²-score Formula

### FÎ²=(1+Î²^2)*(Precision * Recall)/((Î²^2 * Precision)+recall)

```python
y_true = [0,1,1,0,1,0,1,0,0,1]
y_pred = [0,1,0,0,1,1,0,1,0,0]

f1=fbeta_score(y_true,y_pred,beta=1)
f05=fbeta_score(y_true,y_pred,beta=0.5)
f2=fbeta_score(y_true,y_pred,beta=2)


print(f"F1:{f1},F0.5:{f05},F2:{f2}")
```
