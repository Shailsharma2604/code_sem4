# Matplotlib Example Workflow

## 1. Import matplotlib and get it ready for plotting in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2. Prepare Data

```python
x=[1,2,3,4]
y=[11,22,33,44]
```

## 3. Setup plot

```python
fig,ax=plt.subplots(figsize=(5,5))
```

## 4. Plot Data

```python
ax.plot(x,y)
```

## 5. Customize Plot

```python
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## 6. Save a show (you save the valve whole figure)

```python
fig.savefig("./image./sample-plot.png")
```

## 7. Together

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[11,22,33,44]

fig,ax=plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

fig.savefig("./image./sample-plot.png")
```

# Which one should you use ? 
## pyplot vs matplotlib OO method?

### * When plotting something quickly, okay to use the pyplot method
### * When plotting something more advanced, use the OO method

# OO method (Object oriented method)

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50
```

# OO method from scratch

```python
fig,ax=plt.subplots(figsize=(10,6))
```

## Plot the data

```python
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
```

## Customize the plot

```python
ax.set(title="heart disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="cholesterol")
```

## Add a legend

```python
ax.legend(*scatter.legend_elements(),title="Traget")
```

## Add a horixontal line or meanline

```python
ax.axhline(over_50["chol"].mean())
```

## To change the format of line

```python
ax.axhline(over_50["chol"].mean(),
           linestyle="--")
```

## Whole together code

```python
## Plot the data
fig,ax=plt.subplots(figsize=(10,6))

## Customize the plot
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

## Add a legend
ax.set(title="heart disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="cholesterol")

## Add a horixontal line or meanline
ax.legend(*scatter.legend_elements(),title="Traget")

## To change the format of line
ax.axhline(over_50["chol"].mean(),
           linestyle="--")
```

## Code together

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        xlabel="Age",
        ylabel="Cholestrol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Add a title to this figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

# sharex=True

## it means that the subplots will share the same x-axis.

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex=True)

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="Cholestrol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");

#Add a title to this figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
