# Anatomy of a Maplotlib Plot

# Matplotlib Example Workflow

## 1. Import matplotlib and get it ready for plotting in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2. Prepare Data

```python
x=[1,2,3,4]
y=[11,22,33,44]
```

## 3. Setup plot

```python
fig,ax=plt.subplots(figsize=(5,5))
```

## 4. Plot Data

```python
ax.plot(x,y)
```

## 5. Customize Plot

```python
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## 6. Save a show (you save the valve whole figure)

```python
fig.savefig("./image./sample-plot.png")
```

## 7. Together

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[11,22,33,44]

fig,ax=plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

fig.savefig("./image./sample-plot.png")
```

```python
import numpy as np
```

# Line Plot

## Create Some Data

```python
x=np.linspace(0,10,100)
x[:10]
```

## Plot The Data

```python
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

# Scatter Plot

## Used Same data to make a scatter

```python
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
```

## Another scatter plot

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
```

# Bar plot

## Make a plot from dictionary

```python
nut_butter_prices={"Almond butter":10,
                   "peanut butter":8,
                   "Cashew butter":12}

fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Shail's Nut Butter Store", 
       ylabel="Prices($)");
```
