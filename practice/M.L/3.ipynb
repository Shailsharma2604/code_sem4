# Anatomy of a Maplotlib Plot

# Matplotlib Example Workflow

## 1. Import matplotlib and get it ready for plotting in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2. Prepare Data

```python
x=[1,2,3,4]
y=[11,22,33,44]
```

## 3. Setup plot

```python
fig,ax=plt.subplots(figsize=(5,5))
```

## 4. Plot Data

```python
ax.plot(x,y)
```

## 5. Customize Plot

```python
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## 6. Save a show (you save the valve whole figure)

```python
fig.savefig("./image./sample-plot.png")
```

## 7. Together

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[11,22,33,44]

fig,ax=plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

fig.savefig("./image./sample-plot.png")
```

# Bar plot

## Make a plot from dictionary

```python
nut_butter_prices={"Almond butter":10,
                   "peanut butter":8,
                   "Cashew butter":12}

fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Shail's Nut Butter Store", 
       ylabel="Prices($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

# Two options for subplots

# Subplots option 1

### Plot to each different axis

#### ax1.bar(nut_butter_prices.keys(), nut_butter_prices.values());
#### ax2.scatter(x, np.sin(x))
#### ax3.plot(x, x/2)
#### ax4.hist(np.random.rand(1000));

```python
fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows = 2,
                                            ncols= 2,
                                            figsize=(10,5))
# ax1.bar(nut_butter_prices.keys(), nut_butter_prices.values());
# ax2.scatter(x, np.sin(x))
# ax3.plot(x, x/2)
# ax4.hist(np.random.rand(1000));
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

# Subplots option 2

```python
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
