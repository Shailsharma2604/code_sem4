# Matplotlib Example Workflow

## 1. Import matplotlib and get it ready for plotting in Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## 2. Prepare Data

```python
x=[1,2,3,4]
y=[11,22,33,44]
```

## 3. Setup plot

```python
fig,ax=plt.subplots(figsize=(5,5))
```

## 4. Plot Data

```python
ax.plot(x,y)
```

## 5. Customize Plot

```python
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
```

## 6. Save a show (you save the valve whole figure)

```python
fig.savefig("./image./sample-plot.png")
```

## 7. Together

```python
%matplotlib inline
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[11,22,33,44]

fig,ax=plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

fig.savefig("./image./sample-plot.png")
```
